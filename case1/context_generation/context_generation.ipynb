```python
def split_text_v2(text):
    # Splitting based on the "Word pairs" keyword
    parts = text.split("Word pairs:")
    british_american = parts[0].split("American:")
    british = british_american[0].replace("British:", "").strip()
    american = british_american[1].strip() if len(british_american) > 1 else ""
    word_pair = parts[1].strip() if len(parts) > 1 else ""
    return british, american, word_pair
# Applying the function to the DataFrame
new_sample_with_q[['british_question', 'american_question', 'word_pairs']] = new_sample_with_q['rewritten_questions'].apply(lambda x: pd.Series(split_text_v2(x)))
```
