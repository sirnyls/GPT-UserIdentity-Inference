```python
import pandas as pd
import json
source_data = pd.read_csv('global_opinions.csv')
```

```python
global_opinions = source_data[(source_data['selections'].str.contains('United States')) & 
                   ((source_data['selections'].str.contains('Britain'))   )]
global_opinions['source'].value_counts()
```

```python
import ast

global_opinions['value_us'] = ''
global_opinions['value_uk'] = ''
global_opinions['index_us'] = ''
global_opinions['index_uk'] = ''
global_opinions['answer_us'] = ''
global_opinions['answer_uk'] = ''

for index, row in global_opinions.iterrows():
    s = row.selections
    trimmed_string = s[len("defaultdict(<class 'list'>, "):-1]
    result_dict = ast.literal_eval(trimmed_string)
    
    # Get hightest value out of dict for US
    value_us = str(max(result_dict['United States']))

    # Get hightest value out of dict for UK
    if 'Britain' in result_dict:
        value_uk = max(result_dict['Britain'])
    elif 'Great Britain' in result_dict:
        value_uk = max(result_dict['Great Britain'])
    
    # Get index of highest value out of answers for US
    index_us = result_dict['United States'].index(max(result_dict['United States']))

    # Get index of highest value out of answers for UK
    if 'Britain' in result_dict:
        index_uk = result_dict['Britain'].index(max(result_dict['Britain']))
    elif 'Great Britain' in result_dict:
        index_uk = result_dict['Great Britain'].index(max(result_dict['Great Britain']))
    
    # Only save the original selections for UK & US and drop other countries
    if 'Britain' in result_dict:
        global_opinions.at[index, 'selections'] = "'United States': "+str(result_dict['United States']) +", 'Britain': "+str(result_dict['Britain'])
    elif 'Great Britain' in result_dict:
        global_opinions.at[index, 'selections'] = "'United States': "+str(result_dict['United States']) +", 'Great Britain': "+str(result_dict['Great Britain'])
    
    # store values in df
    global_opinions.at[index,'value_us'] = value_us
    global_opinions.at[index,'value_uk'] = value_uk
    global_opinions.at[index,'index_us'] = index_us
    global_opinions.at[index,'index_uk'] = index_uk

    # Map index to answer
    options_as_list = ast.literal_eval(global_opinions['options'][index])

    answer_uk = options_as_list[index_uk]
    answer_us = options_as_list[index_us]

    global_opinions.at[index,'answer_uk'] = answer_uk
    global_opinions.at[index,'answer_us'] = answer_us



print(global_opinions)

#print(result_dict)
#print("Index: "+ str(result_dict['United States'].index(max(result_dict['United States'])))+ "; value: "+ str(max(result_dict['United States'])))
#print("Index: "+ str(result_dict['Britain'].index(max(result_dict['Britain'])))+ "; value: "+ str(max(result_dict['Britain'])))
```
