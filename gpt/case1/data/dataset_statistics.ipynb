```python
df_merged =  pd.concat([df_global_opinions, df_gpt, df_quiz])
```

```python
### Category Grouping
def replace_category(val):
    if val in ['A. Social values and attitudes', 'L. Family and relationships', 'M. Race and ethnicity', 'I. Gender and LGBTQ',
                'Society & Relationships', 'Family & Relationships', 'Cultural Preferences', 'History', 'Literature']:
        return 'Society, Culture, and Relationships'
    elif val in ['D. Politics and policy', 'Q. Security', 'K. Immigration and migration']:
        return 'Politics and Governance'
    elif val in ['C. Science and technology','H. Internet and technology', 'J. News habits and media', 'P. Methodological research',
                'O. Regions and countries', 'E. Demographics', 'Education & Knowledge','Technology',
                'General Knowledge Trivia', 'Education']:
        return 'Media, Technology, and Education'
    elif val in ['Entertainment & Arts', 'Food & Drink', 'Holidays & Celebrations', 'Fashion & Style',
                'Work & Career', 'Travel & Commute', 'Housing & Lifestyle', 'Food,Family & Relationships', 
                'Daily Activities', 'Entertainment', 'Music & Arts', 'Movies & TV Shows', 'Animals', 'Travel']:
        return 'Lifestyle, Entertainment, and Daily Living'
    elif val in ['N. Economy and work', 'G. International affairs', 'B. Religion and spirituality', 
                'Spirituality & Religion', 'Environment & Nature', 'Health & Wellness', 'F. Generations and age',
                'Geography', 'Sports & Recreation']:
        return 'Economy, Health, and Environment'


# Apply the function to the column
df_merged['category_group'] = df_merged['category'].apply(replace_category)

df_global_opinions['category_group'] = df_global_opinions['category'].apply(replace_category)
df_gpt['category_group'] = df_gpt['category'].apply(replace_category)
df_quiz['category_group'] = df_quiz['category'].apply(replace_category)


```

```python
print('All: ' + str(df_merged['category_group'].value_counts().sort_index()))
print('GPT: ' +str(df_gpt['category_group'].value_counts().sort_index()))
print('Quiz: ' + str(df_quiz['category_group'].value_counts().sort_index()))
print('GlobalQA: ' + str(df_global_opinions['category_group'].value_counts().sort_index()))


```
